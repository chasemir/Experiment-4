```python
import pandas as pd
```

```python
# load csv file
board = pd.read_csv('board2.csv')
board
```

```python
Instru = board[(board['Track'] == 'Instrumentation') &
(board['Electronics'] > 70) &
(board['Hometown'] == 'Luzon')]

Instru
```

```python
Average = board[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)

Mindy = board[(board['Hometown'] == 'Mindanao') &
(board['Gender'] == 'Female') &
(Average >= 55)]

Mindy
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette('pastel')
```

```python
Average = board[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)

plt.figure( figsize = (10,6))
sns.violinplot(x ='Track', y ='Average', data = board)
plt.title('Average Score by Track')
plt.show()

plt.figure(figsize = (8,6))
sns.barplot(x ='Gender', y ='Average', data = board)
plt.title('Average Score by Gender')
plt.show()

plt.figure(figsize = (8,6))
sns.lineplot(x ='Hometown', y ='Average', data = board)
plt.title('Average Score by Hometown')
plt.show()
```
